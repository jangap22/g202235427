```python
import numpy as np
import pandas as pd

df = pd.read_csv("diabetes.csv")
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

y = df['Outcome']
X = df.drop('Outcome', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체를 생성합니다.
scaler = StandardScaler()

# 입력 데이터 X를 표준화합니다.
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Defining the model 

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam

model = Sequential()

model.add(Dense(20,input_shape=(8,),activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(Adam(lr=0.04),'binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
#fitting the model and predicting 

# fit network
model_history = model.fit(x=X_train, y=y_train, epochs=100, batch_size=3, validation_data= (X_test, y_test))
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test)
y_pred_class = np.argmax(y_pred)
```

```python
# 모델 성능 평가
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test loss:", test_loss)
print("Test accuracy:", test_acc)
```
