```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_linnerud

linnerud = load_linnerud()
df = pd.concat([pd.DataFrame(linnerud.data, columns=linnerud.feature_names),
                pd.DataFrame(linnerud.target, columns=linnerud.target_names)],
               axis=1)
df.tail()
```

```python
y = df["Pulse"]
X = df.drop('Pulse', axis=1)

y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.1, random_state=42)

X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# 기본 학습률로 Adam 옵티마이저 객체 생성
optimizer = Adam()

# 학습률 조정
new_learning_rate = 0.001  # 조정하고자 하는 학습률 값
optimizer.learning_rate.assign(new_learning_rate)

model = Sequential()
model.add(Dense(20, input_shape = (5,)))
model.add(Dense(10))
model.add(Dense(10))
model.add(Dense(1)) 
model.compile(optimizer = optimizer, loss = 'mse', metrics = ['mse'])

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs = 200, validation_data=(X_test, y_test))
```

```python
model.evaluate(X_test, y_test)
```

```python
# 손실(loss)과 평균 제곱 오차(MSE) 추출
loss = history.history['loss']
mse = history.history['mse']

# 그래프 그리기
epochs = range(1, len(loss) + 1)

plt.plot(epochs, loss, 'b', label='Loss')
plt.plot(epochs, mse, 'r', label='MSE')
plt.title('Training Loss and MSE')
plt.xlabel('Epochs')
plt.ylabel('Loss / MSE')
plt.legend()
plt.show()
```
